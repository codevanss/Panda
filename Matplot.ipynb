```python
import matplotlib.pyplot as plt

#Sample Data
x = [1,2,3,4,5]
y = [10,20,25,40,50]

#Creating a Simple Line Plot
plt.plot(x,y)

#Adding Label and Title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

#Show the Plot
plt.show()
```

```python
import numpy as np

x = np.arange(-10,11,1)
y = x**2

#Creating a Simple Line Plot
plt.plot(x,y , marker="o" , linestyle=":" , color = "g" , label = "y = x^2")

#Add Grid
plt.grid(True)
#Adding Label and Title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Y = X^2")
plt.legend()

#Show the Plot
plt.show()
```
