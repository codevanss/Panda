```python
import matplotlib.pyplot as plt

#Sample Data
x = [1,2,3,4,5]
y = [10,20,25,40,50]

#Creating a Simple Line Plot
plt.plot(x,y)

#Adding Label and Title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

#Show the Plot
plt.show()
```

```python
import numpy as np

x = np.arange(-10,11,1)
y = x**2

#Creating a Simple Line Plot
plt.plot(x,y , marker="o" , linestyle=":" , color = "g" , label = "y = x^2")

#Add Grid
plt.grid(True)
#Adding Label and Title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Y = X^2")
plt.legend()

#Show the Plot
plt.show()
```

## Q-- Generate a scatter plot with 20 random points (0, 50) for both x and y.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.random.randint(0,50,20)
y=np.random.randint(0,50,20)

plt.scatter(x=x , y=y , color='blue' , marker='o')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Random Scatter Plot")
plt.grid(True)
plt.show()
```

## Plot a pie chart showing fruit distribution (Apple: 30%, Banana: 25%, Mango: 20%, Grapes: 15%, Orange: 10%)

```python
import matplotlib.pyplot as plt

data = [30,25,20,15,10]
labels = ["Apple" , "Banana" , "Mango" , "Grapes" , "Orange"]
colors= ["red" , "yellow" , "blue","green" , "orange"]
plt.pie(data, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.show()
```

### Q-- Create a histogram for a dataset of 1000 random numbers ( 1 to 100)


```python
import matplotlib.pyplot as plt
import numpy as np

x=np.random.randint(0,101,1000)
plt.hist(x,bins=20,color='blue' , edgecolor='black')
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data")
plt.show()
```

### Q-- Create a filled area plot with y=sin(x)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)

plt.fill_between(x,y,color='skyblue' , alpha=0.5)
plt.plot(x,y,color='blue' , linewidth = 2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Filled Area Plot of sin(x)")
plt.show()
```
